```python
def find_local_maxima(arr, num_peaks=2):
    L = len(arr)
    maxima_indices = []

    for i in range(1, L - 1):
        if arr[i] > arr[i - 1] and arr[i] > arr[i + 1]:
            maxima_indices.append((i, arr[i]))

    maxima_indices.sort(key=lambda x: x[1], reverse=True)
    return [idx for idx, _ in maxima_indices[:num_peaks]]
```

```python
xk_dB = 10 * np.log10(output_buffer)
gold_dB = 10 * np.log10(gold_ref[::-1])
```

```python
peaks_hw = find_local_maxima(xk_dB, num_peaks=2)
peaks_ref = find_local_maxima(gold_dB, num_peaks=2)

print("Peaks in HW output:", peaks_hw)
print("Peaks in Golden Ref:", peaks_ref)
```
